# Explorations data MYH ansökningar 2024 för kurs

## Resultat för ansökningar 2024


```python
# read in resultat ansökningar 2024 csv
import pandas as pd


df_resutalt_ansonkningar = pd.read_excel(
    "../data/2024_kurser/resultat-2024-for-kurser-inom-yh.xlsx",
    sheet_name="Lista ansökningar",
)


df_resutalt_ansonkningar.head(10)
```

```python
# read in inkomna ansökningar 2024

df_inkomna_ansökningar = pd.read_excel(
    "../data/2024_kurser/inkomna-ansokningar-2024-for-kurser.xlsx",
    sheet_name="Lista ansökningar",
)

df_inkomna_ansökningar.head()
```

```python
df_resutalt_ansonkningar.columns
```

```python
beviljade = (
    df_resutalt_ansonkningar.groupby("Utbildningsområde")[[
        "Totalt antal beviljade platser",
        "Antal beviljade platser start 2024",
        "Antal beviljade platser start och slut 2024",
        "Antal beviljade platser start 2025",
    ]]
    .sum()
    .reset_index()
    .sort_values(by= "Utbildningsområde", ascending= True)
)

beviljade.index.name = "id"
beviljade.reset_index(inplace=True)
beviljade
```

```python
df_inkomna_ansökningar.shape
```

```python
# do aggregation too get the total apply for places for each edicational area
ansökta_platser_utbildningsområde = (
    df_inkomna_ansökningar.groupby("Sökt utbildningsområde")[
        [
            "Sökt antal platser 2024",
            "Sökt antal platser 2024 (start och avslut 2024)",
            "Sökt antal platser 2025",
        ]
    ]
    .sum()
    .reset_index()
    .sort_values(by="Sökt utbildningsområde", ascending=True)
)

ansökta_platser_utbildningsområde["Total ansökta platser"] = (
    ansökta_platser_utbildningsområde[
        [
            "Sökt antal platser 2024",
            "Sökt antal platser 2024 (start och avslut 2024)",
            "Sökt antal platser 2025",
        ]
    ].sum(axis=1)
)

ansökta_platser_utbildningsområde.index.name = "id"

ansökta_platser_utbildningsområde = ansökta_platser_utbildningsområde.reset_index()
ansökta_platser_utbildningsområde
```

```python
import duckdb
bevilade_avslag = duckdb.query(
    """--sql
    SELECT
        apu."Sökt utbildningsområde",
        apu."Total ansökta platser",
        b."Totalt antal beviljade platser"
    FROM 
        beviljade b 
    LEFT JOIN 
        ansökta_platser_utbildningsområde apu ON b.id = apu.id;
    """
).df()
```

```python
bevilade_avslag
```

## Graph matplotlib

```python
ansökta_platser_utbildningsområde = ansökta_platser_utbildningsområde.drop("id", axis= 1).set_index("Sökt utbildningsområde")
```

```python
import matplotlib.pylab as plt



fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(
    x= ansökta_platser_utbildningsområde.index,
    height= ansökta_platser_utbildningsområde["Total ansökta platser"]
)


# Remove clutter
ax.spines[["top", "right"]].set_visible(False)
ax.tick_params(axis="x", which= "both", length= 0)


# Contrast
ax.spines[["left", "bottom"]].set_color("#b9bbb6")
ax.tick_params(axis="both", colors= "#a2a49f")



# Description and label -> Call To Action 
ax.set_xticklabels(labels= ansökta_platser_utbildningsområde.index, rotation= 45, ha= "right")
ax.set_title("Antal sökta plaster i olika utbildningsområden år 2024 - 2025 för kurser", fontsize= 12, loc= "left", pad=30, color= "#a2a49f")
ax.set_ylabel("ANTAL SÖKTA PLATSER", loc= "top", rotation= 0, color= "#a2a49f")
ax.yaxis.set_label_coords(0, 1.02)
ax.set_xlabel("UTBILDNIGSOMRÅDE", loc= "left", fontsize= 12, color= "#a2a49f")






fig = ax.get_figure()
fig.tight_layout()
fig.show()
```

```python


def beviljad_plot():

    ax= beviljade.plot(kind= 'bar', figsize=(12,6), width= 0.8)
    ax.set_xticklabels(beviljade.index, rotation= 45, ha= "right")

    fig = ax.get_figure()
    
    return fig, ax



```

```python
import numpy as np

ax = bevilade_avslag.plot(kind="barh", figsize=(12, 6))


# Creating side stacked bar 
length = np.arange(len(bevilade_avslag["Sökt utbildningsområde"]))
ax.barh(
    length, bevilade_avslag["antal_avslag"], label="Ansökt", color="#50628e"
)
ax.barh(
    length,
    bevilade_avslag["Totalt antal beviljade platser"],
    left=bevilade_avslag["antal_avslag"],
    label="Avslag",
    color= "#051f5e"
)


# clutter
ax.legend().remove()
ax.tick_params("y", length= 0)
ax.spines[["top", "right"]].set_visible(False)

# Description and title 
ax.set_yticklabels(bevilade_avslag["Sökt utbildningsområde"])
ax.set_title("Antal avslag och antal bevilade platser", pad= 20)
ax.set_ylabel("UTBILDNINGSOMRÅDE", loc= "top", rotation= 0, fontsize= 12)
ax.yaxis.set_label_coords(0, 1)
ax.set_xlabel("TOTAL AMOUNT", loc= "left", fontsize= 12)


# Annotation
rect1 = ax.patches[0] # Första stapalen (Data/IT)
rect2 = ax.patches[2] # Andra staplen (Ekonomi, admiminstration och försäljning)

# ax.text(
#     rect1.get_width() + 200,
#     rect1.get_y() + rect1.get_height() + 7,
#     "Avslag",
#     va = "center"
# )


# ax.text(
#     rect2.get_width() + 7500,
#     rect2.get_y() + rect2.get_height() + 5,
#     "Bevilajde platser",
#     va= "center"
# )

ax.annotate(
    "Avslag för ansökta platser",
    xy= (rect1.get_width(), rect1.get_y() + 4),
    ha="center",
    va="center",
    arrowprops= dict(arrowstyle= "->", linewidth= 0.8, connectionstyle= "arc3, rad=.1 "),
    xytext = (rect1.get_height() +4000, rect1.get_y() + 6)
    
)

ax.annotate(
    "Beviljade ansökta platser",
    xy= (rect2.get_height() +10000, rect2.get_y() + 3),
    ha= "center",
    va= "center",
    arrowprops=dict(facecolor= "black", shrink= 0.02),
    xytext= (rect2.get_height() + 10000, rect2.get_y() + 6)
)

fig = ax.get_figure()
fig.tight_layout()
fig.show()
```
