### join and union too create a singel df with all data

```python
import duckdb

# using join too create a dataframe with the data from second sheet
df_lista_flera_kommuner = duckdb.query(
    """--sql
    SELECT
        d1.Diarienummer,
        d1."Anordnare namn",
        d1.Utbildningsnamn,
        d1."Sökt utbildningsområde",
        d1."sökt antal platser 2024",
        d1."Sökt antal platser 2024 (start och avslut 2024)",
        d1."sökt antal platser 2025",
        d1."YH-poäng",
        d2.Kommun,
        d2.Län,
        d2."FA-region",
        d1."Antal Kommuner",
        d1."Antal län",
        d1."Antal FA-regioner"
    FROM 
        df_flera_kommuner d1
    LEFT JOIN 
        df_lista_flera_kommuner d2 ON d1.Diarienummer = d2.Diarienummer 
    """
).df()
```

```python

# Union, so the data with kommuner (se lista flera kommuner) is in this one 

df = duckdb.query(
    """--sql
    SELECT
        Diarienummer,
        "Anordnare namn",
        Utbildningsnamn,
        "Sökt utbildningsområde",
        "Sökt antal platser 2024",
        "Sökt antal platser 2024 (start och avslut 2024)",
        "Sökt antal platser 2025",
        "YH-poäng",
        Kommun,
        Län,
        "FA-region",
        "Antal kommuner",
        "Antal län",
        "Antal FA-regioner"
    FROM 
        df
    UNION 
    SELECT 
        Diarienummer,
        "Anordnare namn",
        Utbildningsnamn,
        "Sökt utbildningsområde",
        "Sökt antal platser 2024",
        "Sökt antal platser 2024 (start och avslut 2024)",
        "Sökt antal platser 2025",
        "YH-poäng",
        Kommun,
        Län,
        "FA-region",
        "Antal kommuner",
        "Antal län",
        "Antal FA-regioner"
    FROM 
        df_lista_flera_kommuner
    """
).df()
```
