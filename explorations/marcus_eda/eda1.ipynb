```python
def reading_file():

    dict_courses = {}

    df_kurser_2024 = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/resultat-2024-for-kurser-inom-yh.xlsx"
    )
    df_kurser_2024.drop(
        "Antal beviljade platser start och slut 2024", axis=1, inplace=True
    )
    dict_courses[2024] = df_kurser_2024

    df_kurser_2023 = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/resultat-2023-for-kurser-inom-yh.xlsx"
    )
    dict_courses[2023] = df_kurser_2023

    df_kurser_2022 = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/resultat-2022-for-kurser-inom-yh.xlsx"
    )
    dict_courses[2022] = df_kurser_2022

    df_kurser_2021_lista = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/inkomna-ansokningar-juni-2021-for-korta-utbildningar-kurser-och-kurspaket.xlsx"
    )
    df_kurser_2021_beviljade = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/resultat-juni-2021-for-korta-utbildningar-kurser-och-kurspaket.xlsx"
    )
    df_kurser_2021 = pd.concat([df_kurser_2021_lista, df_kurser_2021_beviljade])
    if "Beslut" in df_kurser_2021:

        df_kurser_2021["Beslut"] = df_kurser_2021["Beslut"].fillna("Avslag")
        dict_courses[2021] = df_kurser_2021
    else:
        dict_courses[2021] = df_kurser_2021

    df_kurser_2020_april = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/Beviljade-korta-utb-kurser-kurspaket-YH-april-2020-1.xlsx"
    )
    df_kurser_2020_juli = pd.read_excel(
        "../../data/anordnare_resultat_kurser_2020-2024/Beviljade-korta-utb-kurser-kurspaket-YH-juli-2020-2.xlsx"
    )
    df_kurser_2020 = pd.concat([df_kurser_2020_april, df_kurser_2020_juli])
    dict_courses[2020] = df_kurser_2020

    return dict_courses


dict_c = reading_file()

dict_c.keys()
```

```python
def filter_data_platser_program(df):
    df_summary = (
        df.groupby("Utbildningsanordnare administrativ enhet")[
            ["Sökta platser totalt", "Beviljade platser totalt"]
        ]
        .sum()
        .reset_index()
    )
    df_summary

    return df_summary
```

```python
def count_perc_program(df_summary):

    df_perc_count = (
        df_summary["Beviljade platser totalt"]
        / df_summary["Sökta platser totalt"]
        * 100
    )

    return df_perc_count


count = filter_data_platser_program(df_program_2024)
count_perc_program(count)
```

```python
import matplotlib.pyplot as plt

plt.style.use("marcus.mplstyle")

fig, ax = plt.subplots(figsize=(9, 4), dpi=200)

ax.plot(
    df_line_chart.index,
    df_line_chart["Kurser"],
    marker="o",
    color="#2A1B64",
    label="Kurser",
    markeredgecolor="#514E5C",
    markevery=[0, -1],
)

ax.plot(
    df_line_chart.index,
    df_line_chart["Program"],
    marker="o",
    color="#A8A89C",
    label="Program",
    markeredgecolor="#A5A2AD",
    markevery=[0, -1],
)

ax.set_xticks(df_line_chart.index)
ax.set_xlim(df_line_chart.index.min())


# clutter
ax.tick_params("both", color= "#9B9696")

# Description and lable -> call to action
ax.set_xlabel("# ÅR", loc="left", fontsize=8, color= "#657072")
ax.set_ylabel("PROCENT %", loc="top", rotation=0, fontsize= 8, color= "#657072")
ax.yaxis.set_label_coords(-0.01, 1.02)
ax.set_title(
    "Kraftigt ökning i beviljande för kurser mellan 2021-2024, \nmen för program vände det nedåt 2024",
    loc="left", color= "#657072"
)


def text_annotation(ax, **option):

    ax.text(
        x=option.get("x_value", 2022),
        y=option.get("y_value", 30),
        s=option.get("text", ""),
        fontsize=option.get("font", 8),
        color= option.get("color", "grey")
    )


# Text annotation

text_annotation(
    ax, x_value=2024, y_value=52.3, text="51.2% Kurser", font=8
)
text_annotation(
    ax, x_value=2024, y_value=28.2, text="27% Program", font=8
)
text_annotation(
    ax, x_value=2021 + 0.08, y_value=20, text="14%", font=8
)
text_annotation(
    ax, x_value=2021 + 0.08, y_value=37, text="32%", font=8
)
text_annotation(
    ax,
    x_value=2023 - 0.43,
    y_value=20,
    text="Från 2023 till 2024 har program" 
    + "\nen $\mathbf{45,6\%}$ minskning" 
    + "\ni beviljandegrad." 
    + "\nTrots att antal ansökningar" 
    + "\nbara ökat med $\mathbf{1.1\%}$",
    font=8,
)

def arrow_annotation(ax, **option):
    
    x= option.get("x_value", 0)
    y = option.get("y_value_piont", 0)
    y_start = option.get("y_value_start", 0)
    # annotation
    ax.annotate(
        text="",
        xy=(x + option.get("x_distans"), y),  # punkten som pilen ska peka på
        xytext=(x + option.get("x_distans_start", 0), y_start),  # startpunkten för pilen
        arrowprops=dict(arrowstyle="->", linewidth=option.get("linewidth", 0.3), connectionstyle=f"arc3, rad= {option.get('rad', 0.3)}", color= option.get("color", "grey")),
)


arrow_annotation(ax, x_value= 2023, y_value_piont= 40, x_distans= 0.3, y_value_start= 30, linewidth= 0.3, rad= -0.3, color= "grey")
arrow_annotation(ax, x_value= 2021, y_value_piont= 33, x_distans= 0.02, y_value_start= 36.8, x_distans_start= 0.08, linewidth= 0.3, rad= 0.2, color= "grey")
arrow_annotation(ax, x_value= 2021, y_value_piont= 15, x_distans= 0.021, y_value_start= 19, x_distans_start= 0.08, linewidth= 0.3, rad= 0.2, color= "grey")



fig.tight_layout()
fig.show()
fig.savefig("../figures/grant_rate_line_chart.png")
```
